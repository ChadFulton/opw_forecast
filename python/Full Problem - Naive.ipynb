```python
import numpy as np
import pandas as pd
from scipy.misc import comb
from scipy import stats
import matplotlib.pyplot as plt
import re

from __future__ import division
```

```python
# Load the raw data
raw = pd.read_excel('data/recession probit data.xlsx', 'nber_indicator.csv')
raw.index = pd.date_range(start='1960-01-01', end='2011-06-01', freq='MS')
state_cols = raw.columns[8:]
raw.ix[0:5,1:8]
```

```python
# Create the estimation variables
growth = lambda col: (col / col.shift(3) - 1)*100
raw['sp500_return'] = growth(raw['S&P 500'])
raw['term_spread'] = raw['Ten Year Bond'] - raw['Three Month Bill']
raw['agg_emp_growth'] = growth(raw['U.S. Employment'])
raw['agg_ip_growth'] = growth(raw['U.S. Industrial Production'])

# State-level unemployment
state_emp = np.asarray(raw[state_cols])
state_cols = [
    ('%s_emp_growth' % re.sub(' ?\(SA ?, ?[Tt]hous.?\) ?', '', state).lower()).replace(' ', '_')
    for state in state_cols
]
state_cols[27] = '_'.join(state_cols[27].split('_')[1:])
state_cols[49] = state_cols[49].replace('.', '')
state_emp_growth = pd.DataFrame(
    (state_emp[3:] / state_emp[:-3] - 1) * 100,
    index = raw.index[3:],
    columns = state_cols
)
full = pd.merge(raw, state_emp_growth, left_index=True, right_index=True)
```

```python
# Final dataset
agg_cols = ['sp500_return', 'term_spread', 'agg_emp_growth', 'agg_ip_growth']
data = full.ix[1:, ['NBER Indicator', 'Federal Funds Rate'] + agg_cols + state_cols]
data.rename(columns = {'NBER Indicator':'s', 'Federal Funds Rate':'ff'}, inplace=True)
data['intercept'] = np.ones(data.shape[0])
exog_cols = ['intercept', 'ff'] + agg_cols + state_cols
endog_cols = ['s']
# Reorder the columns
data = data.ix[:, endog_cols + exog_cols]

# Since endogenous is next-period recession, shift our exogenous forward one period
data[exog_cols] = data[exog_cols].shift(1)
data = data[1:]

# Create matrices
s = np.asarray(data.ix[:,endog_cols])
x = np.asarray(data.ix[:,exog_cols])

# Constants
T = x.shape[0]
n = x.shape[1]-1
```

### Objects:

$$
\begin{align}
    S_t & \quad \text{Recession=1} &  \in \{0,1\}\\
    h & \quad \text{Forecast lag} & \\
    X_t & \quad \text{Predictors} & n \times 1\\
    \alpha & \quad \text{Intercept} & 1 \times 1 \\
    \beta & \quad \text{Slope} & n \times 1 \\
    \rho & \quad \text{Parameters} & [\alpha, \beta']~\text{is}~n+1 \times 1 \\
    \gamma & \quad \text{Selectors} & \in \{0,1\}^n \\
    k_\gamma & \quad \text{# Selected} & = \rho' \rho \le n \\
    X_\gamma & \quad \text{Included Predictors} & k_\gamma x 1 \\
    \rho_\gamma & \quad \text{Included Parameters} & k_\gamma+1 x 1 \\
    y_t & \quad \text{Data Augmentation} & = \alpha + X_{\gamma, t-h}' \beta_\gamma + u_t~\text{is}~1 \times 1
\end{align}
$$

### Priors

**Parameters**:

$$
p(\rho_\gamma) = N(\mathbf{0}_{k_\gamma+1}, \sigma^2 \mathbf{I}_{k_\gamma+1}); \qquad \sigma^2 = 10
$$

**Selectors**:

"multinomial distribution defined across the $2^n$ different possible choices of $\gamma$", so that the prior probability is:

$$
Pr(\gamma) \propto \frac{1}{{n \choose k_\gamma}}
$$

### Sampling

- **Block 1**: $\rho_\gamma$ and $\gamma$ sampled conditional upon $S^t$ and $y^t = [y_{h+1}, \dots, y_t]'$ as a draw from:

    $$\pi(\rho_\gamma, \gamma \big | y^t)$$

- **Block 2**: Augmented data $y^t$ is sampled conditional on $\rho_\gamma$, $\gamma$, and $S^t$ as a draw from:

    $$\pi(y^t \big | \rho_\gamma, \gamma, S^t)$$

### Data Vectors

#### Augmented data:

$$
y_t = \alpha + X_{\gamma, t-h}' \beta_\gamma + u_t
$$

and

$$
\mathbf{y}^t = \begin{bmatrix}
y_{h+1} \\ \vdots \\ y_t
\end{bmatrix}
$$

is $t-h \times 1$.

#### Observed data:

$$
\mathbf{X}_\gamma^{t-h} = \left ( \mathbf{X}_{\gamma,t-h} \right)_{t=h+1,\dots,t} = \begin{bmatrix}
    \mathbf{X}_{\gamma,(h+1)-h} \\
    \mathbf{X}_{\gamma,(h+2)-h} \\
    \vdots \\
    \mathbf{X}_{\gamma,t-h} \\
\end{bmatrix} = \begin{bmatrix}
    1 & X_{\gamma,1}' \\
    1 & X_{\gamma,2}' \\
    \vdots & \vdots \\
    1 & X_{\gamma,t-h}' \\
\end{bmatrix}
$$

which is $(t-h) \times (k_\gamma + 1)$.

Note that it is just a selection of columns from:

$$
\mathbf{X}^{t-h} = \begin{bmatrix}
    1 & X_{1,1} & X_{2,1} & \cdots & X_{n,1} \\
    1 & X_{1,2} & X_{2,2} & \cdots & X_{n,2} \\
    \vdots & \vdots \\
    1 & X_{1,t-h} & X_{2,t-h} & \cdots & X_{n,t-h} \\
\end{bmatrix}
$$

which is just a selection of rows from:

$$
\mathbf{X} = \begin{bmatrix}
    1 & X_{1,1} & X_{2,1} & \cdots & X_{n,1} \\
    1 & X_{1,2} & X_{2,2} & \cdots & X_{n,2} \\
    \vdots & \vdots \\
    1 & X_{1,t} & X_{2,t} & \cdots & X_{n,t} \\
\end{bmatrix}
$$

### Block 1: Sampling selectors and parameters

```python
def draw_gamma(gamma, rvs):
    """
    Draw a proposal for the selectors vector gamma
    
    Modifies gamma based on the value of `random_int` as follows:
    - If rvs == 0 (which corresponds to the intercept), then gamma_star = gamma
    - Otherwise, gamma_star[rvs] = gamma[random_int]^1
    
    Parameters
    ----------
    gamma : array-like
        The current gamma vector
    rvs : int
        A variate from a discrete uniform distribution over [0,1, ..., n]
        
    Returns
    -------
    gamma_star : array-like
        The drawn (proposed) gamma vector
    """
    gamma_star = np.copy(gamma)
    if rvs > 0:
        gamma_star[rvs] = np.bitwise_xor(gamma_star[rvs], 1)
    return gamma_star
```

```python
def draw_rho(gamma, sigma2, y, x, h, rvs):
    """
    Draw a proposal for the parameters vector rho
    
    Note that because the MH acceptance probability only depends upon
    the proposed gamma vector, this only needs to be called in the event
    that the proposal will be accepted.
    
    Parameters
    ----------
    gamma : array-like
        Selection vector
    m0 : array-like
        Prior mean (note that there are n+1 possible priors)
    M0 : array-like
        Prior covariance matrix (note that there are n+1 possible priors)
    y : array-like
        Augmented data
    x : array-like
        Observed data
    h : integer
        Forecast lag
    rvs : array-like
        Variates from a multivariate (:math:`k_gamma+1 \times 1`) standard normal distribution
        
    Returns
    -------
    rho_star : array-like
        The drawn (proposed) rho vector
    """
    # Data
    k_gamma = gamma.sum()-1
    exog = x[:-h, gamma]
    endog = y
    
    # Posterior covariance matrix
    M1 = np.linalg.inv(np.eye(k_gamma+1)/sigma2 + exog.T.dot(exog)) # TODO can probably prefetch all of these
    # (actually probably cache them - a better idea since there are 2^N of them...)
    # (this will work if there are a selection of gammas that are hit relatively often)
    # (will likely want to have some kind of cache expiration as well)
    
    # Posterior mean
    m1 = M1.dot(exog.T.dot(endog))
    
    # Transform the variate from standard normal to the actual posterior
    A = np.linalg.cholesky(M1) # TODO can probably prefetch/cache these too
    rho_star = m1 + A.dot(rvs[:k_gamma+1])
    return rho_star
```

```python
def mvn_density(gamma, sigma2, endog, exog):
    Sigma = np.eye(exog.shape[0]) + sigma2*exog.dot(exog.T)
    return (
        (np.linalg.det(Sigma)**0.5) *
        np.exp(-0.5*endog.T.dot(np.linalg.inv(Sigma)).dot(endog))
    )

def mn_mass(gamma):
    return 1/comb(gamma.shape[0], gamma.sum())

def accept_prob(gamma, gamma_star, sigma2, y, x, h):
    endog = y
    
    return (
        (mvn_density(gamma, sigma2, endog, x[:-h,gamma]) * mn_mass(gamma)) /
        (mvn_density(gamma_star, sigma2, endog, x[:-h,gamma_star]) * mn_mass(gamma_star))
    )
```

### Block 2: Sampling augmented data

```python
def draw_y(rho, gamma, s, x, h):
    T = s.shape[0]
    xB = np.dot(x[:-h,gamma], rho[gamma])
    y = np.zeros((T-h,))
    for t in range(T-h):
        if s[t] == 1:
            y[t] = stats.truncnorm.rvs(-xB[t], np.Inf, loc=xB[t])
        else:
            y[t] = stats.truncnorm.rvs(-np.Inf, -xB[t], loc=xB[t])
    return y
```

## Metrpolis-within-Gibbs Sampling

```python
%%prun
# Priors
sigma2 = 10
h = 1

# Parameters
G0 = 0
G = 100

# Data arrays
gammas = np.ones((n, G0+G+1), dtype=bool)
#gammas[0,:] = True
rhos = np.zeros((n+1, G0+G+1))
ys = np.zeros((T-h, G0+G+1))

# Random variables
comparators = np.random.uniform(size=(G0+G))
gamma_rvs = np.random.random_integers(0, n-1, size=(G0+G,))
rho_rvs = np.random.multivariate_normal(mean=np.zeros(n+1), cov=np.eye(n+1)*sigma2, size=(G0+G,))

# Iterations
for t in range(1, G0+G+1):
    # 1. Metropolis step: draw gamma and rho
    gamma = gammas[:, t-1].copy()
    gamma_star = draw_gamma(gamma, gamma_rvs[t-1])
    accept = accept_prob(gamma, gamma_star, sigma2, ys[:,t-1], x, h)
    if accept > comparators[t-1]:
        gamma = gamma_star
        rhos[gamma,t] = draw_rho(gamma, sigma2, ys[:,t-1], x, h, rho_rvs[t-1,:])
    else:
        rho = rhos[:,t-1].copy()
    gammas[:, t] = gamma
    
    # 2. Gibbs step: draw y
    ys[:,t] = draw_y(rhos[:,t], gammas[:,t], s, x, h)
```
